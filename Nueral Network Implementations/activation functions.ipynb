# Relu function
![Alt text](https://www.nomidl.com/wp-content/uploads/2022/04/image-10.png)


```python
import numpy as np

def relu(x):
    return np.maximum(0, x)

# Example usage
x = np.array([-35])
result = relu(x)

print(result)
```

# Sigmoid function
![Alt text](https://miro.medium.com/v2/resize:fit:970/1*Xu7B5y9gp0iL5ooBj7LtWw.png)

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Example usage
x = np.array([100,200])
result = sigmoid(x)

print(result)
```

# Softmax function
![Alt text](https://miro.medium.com/v2/resize:fit:1084/1*PFAEF4oRm_2nf759NAweqQ.png)


Alt text

```python
import numpy as np

def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))  # Subtracting max(x) for numerical stability
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)

# Example usage
x = np.array([1.0, 2.0, 3.0])
result = softmax(x)

print(result)
```
